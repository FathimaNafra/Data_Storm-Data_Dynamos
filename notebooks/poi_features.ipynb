```python
import osmnx as ox
import pandas as pd
```

```python
coordinates = pd.read_csv(
    "../data/silver/clean_outletCoordinates.csv"
)
```

```python
schools = ox.features_from_place(
    "Colombo, Sri Lanka",
    tags={"amenity": "school"}
)
```

```python
#Load gold features and merge with coordinates
features = pd.read_csv("../data/gold/features.csv")
features = features.merge(
    coordinates,
    left_on="gitOutlet_ID",
    right_on="Outlet_ID",
    how="left"
)
```
